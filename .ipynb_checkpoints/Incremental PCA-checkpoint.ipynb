```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.decomposition import PCA, IncrementalPCA
```

```python
data = pd.read_csv('e:/SklearnTest/originalData/L-cal.csv')
X = data.iloc[:,0:396]
y = data.iloc[:,-1]
```

```python
y.value_counts()
```

```python
n_components = 3
ipca = IncrementalPCA(n_components=n_components,batch_size = 10)
X_ipca = ipca.fit_transform(X)

pca = PCA(n_components=n_components)
X_pca = pca.fit_transform(X)
```

```python
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
fig = plt.figure()
ax = Axes3D(fig, rect=[0, 0, 2, 2], elev=30, azim=20)
colors = ['navy', 'turquoise', 'darkorange','red']

for X_transformed, title in [(X_ipca,'Incremental PCA'),(X_pca,'PCA')]:
    plt.figure(figsize=(8,8))
    for color, i, target_name in zip(colors,[0,1,2,3,4],y):
        ax.scatter(X_transformed[y==i,0],X_transformed[y==i,1],
                   color=color,lw=2,label = target_name)
plt.show()
```

```python
print(ipca.explained_variance_ratio_,'\n',pca.explained_variance_ratio_)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
data_pre = pd.read_csv('e:/SklearnTest/originalData/L-pre.csv')
X_p = data_pre.iloc[:,0:396]
y_p = data_pre.iloc[:,-1]

X_pipca = ipca.fit_transform(X_p)
```

```python
LR = LogisticRegression()
LR.fit(X_ipca,y)
y_pred = LR.predict(X_pipca)
```

```python
from sklearn import metrics

print('>>>在训练集的表现',LR.score(X_ipca,y))
print('>>>在测试集的表现',metrics.accuracy_score(y_pred,y_p))
print('>>>混淆矩阵\n',metrics.confusion_matrix(y_p,y_pred))
print('>>>分类评价函数\n',metrics.classification_report(y_p,y_pred))
# y_predprob = LR.predict_proba(X_pipca)[:,1]  
# print ("AUC Score (Train): %f" % metrics.roc_auc_score(y, y_predprob))  #不支持多分类
```

```python
from sklearn.preprocessing import scale
X_sc = scale(X)
X_psc = scale(X_p)
```

```python
#X_ppca = pca.fit_transform(X_p)#n_components='mle' is only supported if n_samples >= n_features
pca = PCA(n_components=250)
X_pca = pca.fit_transform(X_sc)
X_ppca = pca.fit_transform(X_psc)
```

```python
X_ppca.shape
```

```python
LR = LogisticRegression()
LR.fit(X_pca,y)
y_pred = LR.predict(X_ppca)
from sklearn import metrics

print('>>>在训练集的表现',LR.score(X_pca,y))
print('>>>在测试集的表现',metrics.accuracy_score(y_pred,y_p))
print('>>>混淆矩阵\n',metrics.confusion_matrix(y_p,y_pred))
print('>>>分类评价函数\n',metrics.classification_report(y_p,y_pred))
```
